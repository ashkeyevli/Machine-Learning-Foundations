# First task

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['word_count'] = turicreate.text_analytics.count_words(products['review'])
```

```python
products.show()
```

```python
products
```

```python
for word in selected_words:
    products[word] = products['word_count'].apply(lambda selected_words_count: selected_words_count[word] if word in selected_words_count else 0)
```

```python
products
```

# Test some feaures

```python
def awesome_count(words_count):
    b = {}
    if 'awesome' in words_count:
        b['awesome'] = words_count['awesome']
    else:
        b['awesome'] = 0
    return b['awesome']
```

```python
products['awesome2'] = products['word_count'].apply(awesome_count)
```

```python
products
```

```python
max_awesome2 = products[products['awesome2']>3]
```

```python
max_awesome2
```
