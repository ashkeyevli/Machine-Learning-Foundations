# First task

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['word_count'] = turicreate.text_analytics.count_words(products['review'])
```

```python
products.show()
```

```python
products
```

```python
for word in selected_words:
    products[word] = products['word_count'].apply(lambda selected_words_count: selected_words_count[word] if word in selected_words_count else 0)
```

```python
products
```

# Test some feaures

```python
def awesome_count(words_count):
    b = {}
    if 'awesome' in words_count:
        b['awesome'] = words_count['awesome']
    else:
        b['awesome'] = 0
    return b['awesome']
```

```python
products['awesome2'] = products['word_count'].apply(awesome_count)
```

```python
products
```

```python
max_awesome2 = products[products['awesome2']>3]
```

```python
max_awesome2
```

# Second question MOST and LEAST 


```python
for word in selected_words:
    products[word + '2'] = products[word].sum()
```

```python
products
```

# Second task


```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
features=selected_words
```

```python
#ignore all 3*  reviews
products = products[products['rating']!= 3]
```

```python
#positive sentiment = 4-star or 5-star reviews
products['sentiment'] = products['rating'] >= 4
```

```python
products
```

```python
sentiment_selected_model = turicreate.logistic_classifier.create(train_data,target='sentiment', features=selected_words, validation_set=test_data)
```

```python
sentiment_selected_model
```

```python
coefficients = sentiment_selected_model.coefficients
```

```python
coefficients
```

```python
sentiment_selected_model.coefficients
```

```python
coef2 = coefficients.sort('value', ascending=False)
```

```python
coef2[0]
```

```python
coef2[-1]
```

```python
coef2.tail()
```

```python
evolutaion_results = sentiment_selected_model.evaluate(test_data, metric = 'roc_curve')
```

```python
sentiment_selected_model.evaluate(test_data)
```

```python
evolutaion_results
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(evolutaion_results['roc_curve']['fpr'], evolutaion_results['roc_curve']['tpr'])
```
